```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df_1 = pd.read_csv('train.csv')
df_2 = pd.read_csv('test.csv')
```

```python
df_com = pd.concat([df_1, df_2], ignore_index=True)
```

```python
df_com
```

```python
label_col = 'fake'
features = df_com.drop(columns=[label_col])
label = df_com[label_col]

scaler = StandardScaler()
scaler.fit(features)
normalized_features = scaler.transform(features)
df_normalized = pd.DataFrame(normalized_features, columns=features.columns)

df_normalized[label_col] = label
df_normalized.to_csv('comb_Z_score.csv', index=False)

print("완료")
```
